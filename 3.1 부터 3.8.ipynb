```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
a=np.array([4,5,6])

fig = plt.figure(figsize=plt.figaspect(0.5)*1.5)

ax=fig.add_subplot(1,1,1,projection='3d')

ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0],a[1],a[2],'a',size=15)

ax.set_xlim(0,8)
ax.set_ylim(0,8)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20.,azim=5)
ax.grid()
ax.set_axisbelow(True)

ax = fig.gca(projection='3d');
```

```python
a=np.array([4,3])

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

ax.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1)
ax.text(a[0],a[1],'a',size=15)

ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero');
```

벡터의 스칼라배(양수,음수)

```python
a=np.array([4,1])
c=3
ca=np.multiply(c,a)

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1)
plt.text(a[0],a[1],'a',size=15)

plt.quiver(0,1,ca[0],ca[1],angles='xy',scale_units='xy',scale=1,color='blue')
plt.text(ca[0],ca[1],'ca',size=15,color='blue')

ax.set_xticks(range(-2,15))
ax.set_yticks(range(-3,9))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none');
```

```python
a=np.array([4,1])
c=-3
ca=np.multiply(c,a)

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1)
plt.text(a[0],a[1],'a',size=15)

plt.quiver(0,1,ca[0],ca[1],angles='xy',scale_units='xy',scale=1,color='red')
plt.text(ca[0],ca[1],'-3a',size=15,color='red')

ax.set_xticks(range(-10,7))
ax.set_yticks(range(-5,5))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none');
```
